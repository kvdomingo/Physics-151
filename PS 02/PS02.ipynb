```python
import numpy as np
import matplotlib.pyplot as mp
import scipy.stats as stat
from jupyterthemes import jtplot
```

```python
jtplot.reset()
mp.rcParams["text.usetex"] = True
mp.rcParams["font.family"] = "serif"
mp.rcParams["figure.figsize"] = (3*16/9, 3)
mp.rcParams["figure.dpi"] = 100
```

```python
N = np.array([40, 60, 80, 100], int)
Ed_40 = np.array([.655, .436, .331, .265])
Ed_80 = np.array([1.308, .879, .66, .532])
E_40 = np.array([39.345, 39.564, 39.669, 39.735])
E_80 = np.array([78.692, 79.121, 79.34, 79.468])
EN_40 = E_40/N
EN_80 = E_80/N
```

```python
m, b, R, p, sd = stat.linregress(Ed_40, EN_40)
mp.plot(Ed_40, EN_40, "bo")
mp.plot(Ed_40, m*Ed_40+b, "b--", label=r"$\langle E_d \rangle \langle E \rangle^{-1} N = %.3f$"%m)

m, b, R, p, sd = stat.linregress(Ed_80, EN_80)
mp.plot(Ed_80, EN_80, "go")
mp.plot(Ed_80, m*Ed_80+b, "g--", label=r"$\langle E_d \rangle \langle E \rangle^{-1} N = %.3f$"%m)

mp.xlabel(r"$\langle E_d \rangle$")
mp.ylabel(r"$\langle E \rangle /N$")
mp.legend()
mp.grid()

mp.savefig("demonenergy.png", dpi=300, bbox_inches="tight")
mp.show()
```
