```python
x = np.logspace(1, 3, 500, dtype="float64", base=10)
y = np.zeros_like(x)

for k in range(len(x)):
    y[k] = gamma(x[k]-1+20-1)/gamma(x[k]-1)/gamma(19-1)

fig = mp.figure()
ax = fig.add_subplot(111)
ax.plot(x, y, "k-")
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("$E$")
ax.set_ylabel("$\Omega(E)$")
ax.grid(True)

mp.savefig("microstates.png", dpi=300, bbox_inches="tight")
mp.show()
```

```python
x = np.arange(20, 101, 1, dtype="float64")
y = np.zeros_like(x)

for k in range(len(x)):
    y[k] = gamma(10-1+x[k]-1)/gamma(10-1)/gamma(x[k]-1-1)

fig = mp.figure()
ax = fig.add_subplot(111)
ax.plot(x, y, "k-")
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("$N$")
ax.set_ylabel("$\Omega(E)$")
ax.grid(True)

mp.savefig("fixedN.png", dpi=300, bbox_inches="tight")
mp.show()
```

```python
def F(E, V, N):
    return (V/N)**N * (E/3/N/np.pi/hbar**2)**(3*N/2) * np.exp(5*N/2)
```

```python
E = np.logspace(1, 3, 500, base=10)
V = np.linspace(1, 1000, 500)
N = np.round(np.linspace(10, 1000, 500))

mp.plot(E, F(E,V,N), "k-")
mp.xscale("log")
mp.yscale("log")
mp.grid(True)
mp.show()
```
